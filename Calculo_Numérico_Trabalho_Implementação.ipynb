# Metodo gráfico para determinar o intervalo do zero da função

Aqui será utilizado o que no livro é conhecido como isolamento das raízes e seguindo assim a Fase 1

Vamos plotar o gráfico da função f(x) para determinar o intervalo que cruza o eixo x pelo processo (i) da pagina 33

```python
x = np.linspace(0,3,100)
y = (5*np.log10(x))-2+(0.4*x)

fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,y,'y', label='y=5*log10(x)-2+0,4*x')
plt.legend(loc='upper left')
plt.show()
```

Conforme o gráfico acima é possivel observar que a f(x) cruza o eixo x no intervalo [ 1.5 , 2 ]

Podemos agora fazer a confirmação pelo processo (ii) da pagina 33 em que se gera duas funções g(x) e h(x) tomando f(x)=0 e esboçar seus graficos a fim de encontrar o ponto de interceptação

```python
x = np.linspace(0,3,100)
h = (5*np.log10(x))
g = (2-(0.4*x))
fig = plt.figure(figsize=(10,8))


ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,h,'h', label='h(x) = 5*log10(x)')
plt.legend(loc='upper left')
plt.plot(x,g,'g', label='g(x) = -2+(0.4*x)')
plt.legend(loc='upper left')

plt.show()
```

Conforme visto acima no gráfico o ponto de encontro está no intervalo [ 1.5 , 2 ]

# Método da Bissecção Intervalo a=1,5 e b=2,0

Baseado no código disponivel no Classroom foi possivel transpor para linguagem python

```python
a = 1.5
b = 2
p = 10**(-15)
k = 100 
controle = 0
```

```python
M = (5*np.log10(a))-2+(0.4*a)
fx = p+1
```

```python
while((controle < k) and (abs(fx) > p)):
  x = (a+b)/2
  fx = (5*np.log10(x))-2+(0.4*x)
  print(f"\n k = {controle} - x = {x} - f(x) = {fx}")

  if(fx*M>0):
    a=x
  else:
    b=x

  controle+=1
```
